```python
%%file bigramcount.py

import nltk
from mrjob.job import MRJob
from mrjob.step import MRStep

class MRWordFrequencyCount(MRJob):

    def mapper_init(self):
        nltk.download('punkt')
        nltk.download('stopwords')
        
    def steps(self):
        return [
            MRStep(
                mapper_init = self.mapper_init,
                mapper = self.mapper_extract_words, 
                reducer = self.reducer_sum_word_counts
            ),
            MRStep(
                reducer = self.reduce_sort_counts
            )
        ]

    def mapper_extract_words(self, _, line):
        data = line.split("\" \"")
        if data[0] == "\"character":
            return ("", 0)
        dict = {}
        distilled_sentence = data[2].lower().replace('\'', '').replace('\n', '').replace('?', '').replace('!', '').replace(',', '').replace('-', ' ').replace('.','').replace("  ", " ")
        for bg in list(nltk.bigrams(distilled_sentence.split())):
            if (dict.get(bg, -1) == -1):
                dict[bg] = 1
            else:
                dict[bg] += 1
        for key in dict.keys():
            yield(key, dict[key])


    #def combine_word_counts(self, word, counts):
    #    yield word, sum(counts)

    def reducer_sum_word_counts(self, key, values):
        yield None, (key, sum(values))


    def reduce_sort_counts(self, _, word_counts):
        for key, val in sorted(word_counts, key=lambda x: x[1], reverse = True)[:20]:
            yield (key, val)

if __name__ == '__main__':
    MRWordFrequencyCount.run()
```
