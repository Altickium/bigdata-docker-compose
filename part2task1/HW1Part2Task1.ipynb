```python
%%file wordcount.py

from mrjob.job import MRJob

class MRSongCount(MRJob):
    def mapper(self, _, line):
        words = line.split("\" \"")

        i = 0
        for word in words[1:]:
            if (i == 0):
                i = 1
                yield (word, 1)
            else:
                i = 0

    def reducer(self, word, values):
        yield(word, sum(values))

if __name__ == "__main__":
    MRSongCount.run()
```
